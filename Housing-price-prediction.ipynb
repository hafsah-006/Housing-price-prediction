```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
from xgboost import XGBRegressor
df_train = pd.read_csv('/kaggle/input/train-and-test/train_set.csv')
df_test = pd.read_csv('/kaggle/input/train-and-test/test_set.csv')
y = df_train['Total_Sales']
X = df_train.drop(['Total_Sales'], axis=1)
numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
num_imputer = SimpleImputer(strategy='mean')
X[numeric_cols] = num_imputer.fit_transform(X[numeric_cols])
categorical_cols = X.select_dtypes(include=['object']).columns
cat_imputer = SimpleImputer(strategy='most_frequent')
X[categorical_cols] = cat_imputer.fit_transform(X[categorical_cols])
X_encoded = pd.get_dummies(X, columns=categorical_cols)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=0)
my_model = XGBRegressor()
my_model.fit(X_train, y_train)
y_pred = my_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(mse)
X2_test = df_test
X2_test[numeric_cols] = num_imputer.transform(X2_test[numeric_cols])
X2_test[categorical_cols] = cat_imputer.transform(X2_test[categorical_cols])
X2_test_encoded = pd.get_dummies(X2_test, columns=categorical_cols)
X2_test_encoded = X2_test_encoded.reindex(columns=X_encoded.columns, fill_value=0)
y_test_pred =my_model.predict(X2_test_encoded)
print("Predictions for test set:", y_test_pred)
submission = pd.DataFrame({'Id': X2_test['Id'], 'Total_Sales': y_test_pred})
print(submission.shape)
submission.to_csv('submission.csv', index=False)
```
